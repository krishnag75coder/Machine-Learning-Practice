```python
import pickle
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
with open ("model_pickle",'wb') as f:
    pickle.dump(model,f)
```

```python
with open('model_pickle','rb') as f:
    mp = pickle.load(f)
```
